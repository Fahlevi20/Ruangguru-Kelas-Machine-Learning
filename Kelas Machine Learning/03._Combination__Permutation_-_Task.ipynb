```python
from itertools import product
from itertools import permutations
from itertools import combinations
```

```python
# Peluang angka 6 dalam 2 dadu

myList = list(product(range(1, 7), repeat = 2))

# Inisialisasi
nS = 0
nA = 0

for val in myList:
  nS+=1              # Counter untuk menghitung Ruang Sample
  if(sum(val)==6):   # SUMIF kalau jumlah dadunya 6
    print(*val)      # Mencetak list/hasil kombinasi
    nA+=1            # Counter untuk menghitung Kemungkinan

print('Banyak Ruang Sample: ', nS)
print('Banyak Kemungkinan: ', nA)
```

```python
# PERMUTATION 1
import itertools

string = "ABC"
P = itertools.permutations(string)

count = 0

for val in P:
  count+=1
  print(*val)

print('Number of Permutations', count)
```

```python
# PERMUTATION 2

values = [1, 2, 3, 4, 5]

P = itertools.permutations(values)

count = 0

for val in P:
  count+=1
  print(*val)
  
print('Number of Permutations', count)
```

```python
# PERMUTATION 3

values = ['Anna', 'Elsa', 'Tya', 'Wijaya']

P = itertools.permutations(values, 3)

count = 0

for val in P:
  count+=1
  print(*val)
  
print('Number of Permutations', count)
```

```python
# COMBINATION 1

string = "ABCDE"

C = itertools.combinations(string, 5)

for val in C:
  print(*val)
```

```python
# COMBINATION 2

values = [1, 2, 3, 4, 5]

C = itertools.combinations(values, 2)

for val in C:
  print(*val)
```

```python
# COMBINATION 2

values = ['Python', 'SQL', 'Tableau', 'Excel']

C = itertools.combinations(values, 3)

for val in C:
  print(val)
```
